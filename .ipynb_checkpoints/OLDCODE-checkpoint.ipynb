```python
import pandas as pd
import numpy as np
import statsmodels.api as sm 
from statsmodels.tsa.stattools import adfuller, coint
import matplotlib.pyplot as plt
import yfinance as yf
import multiprocessing
import sys
sys.path.append("./core/")
from strategy import run_static_strategy
```

```python
#pick1 = "HDFCBANK.NS"
#pick2 = "ICICIBANK.NS" was extremely successful for few years and then their relationship broke down
pick1 = "JSWSTEEL.NS"
pick2 = "TATASTEEL.NS"
stock1 = yf.Ticker(pick1)
stock2 = yf.Ticker(pick2)
data1 = stock1.history(start="2018-01-01", end="2025-5-30")["Close"]
data2 = stock2.history(start="2018-01-01", end="2025-5-30")["Close"]

data = pd.DataFrame({
    pick1: data1,
    pick2: data2
})

correlation = data[pick1].corr(data[pick2])
print("The correlation between the two stocks is " + str(correlation))
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data.index, data[pick1], label=pick1, color="red")
plt.plot(data.index, data[pick2], label=pick2, color="blue")

plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.title(pick1 + " vs " + pick2 + " Stock Prices Over Time")
plt.legend()
plt.show()
```

```python
score, pvalue, _ = coint(data[pick1], data[pick2])
print("Cointegration test p-value:", pvalue)
```

```python
split_date = "2020-03-01"

train_data = data[data.index < split_date]
test_data = data[data.index >= split_date]

X = train_data[pick1]  
Y = train_data[pick2] 
X = sm.add_constant(X)
model = sm.OLS(Y, X).fit()
print(model.summary())
intercept, slope = model.params
regression_line = intercept + slope * data[pick1]
plt.figure(figsize=(12, 6))
plt.scatter(data[pick1], data[pick2], label="Data Points", color="blue", alpha=0.6)
plt.plot(data[pick1], regression_line, label="Regression Line", color="red", linewidth=2)
plt.xlabel(pick1 + " Stock Price")
plt.ylabel(pick2 + " Stock Price")
plt.title("Regression Line")
plt.legend()
plt.show()
```

```python
# === ROLLING REGRESSION ===
regression_window = 120
zscore_window = 30

spread = []
slopes = []
intercepts = []
dates = []

for i in range(regression_window, len(data)):
    x_window = data[pick1].iloc[i - regression_window:i]
    y_window = data[pick2].iloc[i - regression_window:i]

    slope, intercept = np.polyfit(x_window, y_window, 1)
    spread_val = data[pick2].iloc[i] - (intercept + slope * data[pick1].iloc[i])

    spread.append(spread_val)
    slopes.append(slope)
    intercepts.append(intercept)
    dates.append(data.index[i])

# Create aligned spread series
spread_series = pd.Series(spread, index=dates)

# Rolling Z-score on spread
zscore = (spread_series - spread_series.rolling(zscore_window).mean()) / spread_series.rolling(zscore_window).std()
zscore = zscore.dropna()

# Align everything
aligned_data = data.loc[zscore.index]
returns1 = aligned_data[pick1].pct_change().fillna(0)
returns2 = aligned_data[pick2].pct_change().fillna(0)
slopes_series = pd.Series(slopes, index=dates).loc[zscore.index]
```

```python
# Ensure common index across all time series
common_index = zscore.index.intersection(spread_series.index).intersection(returns1.index).intersection(returns2.index).intersection(slopes_series.index)

# Align all series to the common index
zscore = zscore.loc[common_index]
spread_series = spread_series.loc[common_index]
returns1 = returns1.loc[common_index]
returns2 = returns2.loc[common_index]
slopes_series = slopes_series.loc[common_index]

# Filter to test period (post-COVID)
test_mask = zscore.index >= split_date
zscore_test = zscore[test_mask]
spread_test = spread_series[test_mask]
returns1 = returns1[test_mask]
returns2 = returns2[test_mask]
slopes_series = slopes_series[test_mask]

# Initialize
COST = 0.0005
position = 0
entry_day = -10000
positions = []
pnl = []
prev_position = 0

z_vals = zscore_test.values
r1_vals = returns1.values
r2_vals = returns2.values
slopes_vals = slopes_series.values
index = zscore_test.index

# Backtest loop
for i in range(len(z_vals)):
    z = z_vals[i]
    slope_now = slopes_vals[i]

    # Entry logic
    if position == 0:
        if z < -1:
            position = 1
            entry_day = i
        elif z > 1:
            position = -1
            entry_day = i

    # Exit logic
    elif position > 0 and z > -0.5 and i - entry_day > 2:
        position = 0
    elif position < 0 and z < 0.5 and i - entry_day > 2:
        position = 0

    positions.append(position)

    # PnL
    if i == 0:
        pnl.append(0)
    else:
        daily_return = r2_vals[i] - slope_now * r1_vals[i]
        cost = COST * abs(position - prev_position) if position != prev_position else 0
        pnl.append(position * daily_return - cost)

    prev_position = position
```

```python
results = pd.DataFrame({
    "Z-Score": zscore_test,
    "Position": positions,
    "Strategy Return": pnl
}, index=index)

results["Cumulative Return"] = (1 + results["Strategy Return"]).cumprod()

plt.figure(figsize=(12, 6))
results["Cumulative Return"].plot()
plt.title("Cumulative Return (Rolling Regression Strategy)")
plt.xlabel("Date")
plt.ylabel("Cumulative Return")
plt.grid(True)
plt.show()
```

```python
spread_train = train_data[pick2] - (intercept + slope * train_data[pick1])
spread_mean = spread_train.mean()
spread_std = spread_train.std()

# === Apply to test data ===
spread_test = test_data[pick2] - (intercept + slope * test_data[pick1])
zscore_test = (spread_test - spread_mean) / spread_std
zscore_test = zscore_test.dropna()


returns1 = test_data[pick1].pct_change().fillna(0).loc[zscore_test.index]
returns2 = test_data[pick2].pct_change().fillna(0).loc[zscore_test.index]


COST = 0.0005
position = 0
entry_day = -10000
positions = []
pnl = []
prev_position = 0

z_vals = zscore_test.values
r1_vals = returns1.values
r2_vals = returns2.values
index = zscore_test.index

for i in range(len(z_vals)):
    z = z_vals[i]

    # Entry logic
    if position == 0:
        if z < -1:
            position = 1
            entry_day = i
        elif z > 1:
            position = -1
            entry_day = i

    # Exit logic
    elif position > 0 and z > -0.5 and i - entry_day > 2:
        position = 0
    elif position < 0 and z < 0.5 and i - entry_day > 2:
        position = 0

    positions.append(position)

    # PnL
    if i == 0:
        pnl.append(0)
    else:
        daily_return = r2_vals[i] - slope * r1_vals[i]  # using fixed slope from training
        cost = COST * abs(position - prev_position) if position != prev_position else 0
        pnl.append(position * daily_return - cost)

    prev_position = position


results = pd.DataFrame({
    "Z-Score": zscore_test,
    "Position": positions,
    "Strategy Return": pnl
}, index=index)

results["Cumulative Return"] = (1 + results["Strategy Return"]).cumprod()

plt.figure(figsize=(12, 6))
results["Cumulative Return"].plot()
plt.title("Cumulative Return (Static Model)")
plt.xlabel("Date")
plt.ylabel("Cumulative Return")
plt.grid(True)
plt.show()

```
